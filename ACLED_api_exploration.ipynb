```python
acled_key = input('ACLED Key: ')
#cde_key = input('CDE Key: ')
email = 'aseemang@terpmail.umd.edu'
```

```python
#fields=field|field|field to only keep certain ones,

#This data call returns the main dataset https://api.acleddata.com/acled/read?
acled = requests.get(f'https://api.acleddata.com/acled/read?key={acled_key}&email={email}&country=united states&limit=1&fields=event_date|sub_event_type|assoc_actor_1|assoc_actor_2|admin1|location|notes')
```

```python
#This data call returns the actors https://api.acleddata.com/actor/read/?

actor = requests.get(f'https://api.acleddata.com/actor/read?key={acled_key}&email={email}&limit=5')
```

```python
actor.json()
```

```python
#This data call returns the actor types https://api.acleddata.com/actortype/read?

actor_types = requests.get(f'https://api.acleddata.com/actortype/read?key={acled_key}&email={email}&limit=1')
```
